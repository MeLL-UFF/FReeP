```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv('sciphy.csv')
data = data[~data['erro']].copy().drop('erro', axis=1).reset_index(drop=True)
sns.set(font_scale=.7)


dummie_data = pd.get_dummies(data)
correlation = dummie_data.corr()
labels = correlation.columns
correlation_matrix = sns.heatmap(correlation)
fig = correlation_matrix.get_figure()
fig.savefig("scipy_correlation.pdf",bbox_inches='tight', pad_inches=.1)
```
