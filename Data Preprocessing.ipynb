```python
import pandas as pd
data = pd.read_csv('results/old/categorical_results.csv', float_precision='round_trip', sep=';')
model1 = data[data['FEATURE'] == 'model1'].drop('FEATURE', axis=1)
# OLD TRANSFORM
# time_data = model1.pivot(index='PARTITIONER', columns='CLASSIFIER', values='TIME')
# acc_data = model1.pivot(index='PARTITIONER', columns='CLASSIFIER', values='ACCURACY')
```

```python
import seaborn as sns
sns.set(style="whitegrid")

# Draw a nested barplot to show survival for class and sex
graph = sns.catplot(x="PARTITIONER", y="TIME", hue="CLASSIFIER", data=model1,
                height=6, kind="bar", palette="muted")
graph.despine(left=True)
graph.set_ylabels("TIME")
graph.fig.savefig("test.pdf")
```
